```python
import os
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def sort_df(df):

    cols = [x*100 for x in range(1,501)]
    df.columns = cols

    for col in cols:
        df[col] = df[col].apply(lambda x: x.replace("'", '') if x==x else '')
        df[col] = df[col].apply(lambda x: float(x) if x==x else '')

    avgs = df.mean(axis=0)
    df.loc['averages'] = pd.Series(avgs)
    df = df.transpose()
    return df
```

```python
random_df = sort_df(pd.read_csv("output_tables/random_rewards_per_100.csv"))
simple_df = sort_df(pd.read_csv("output_tables/simple_rewards_per_100.csv"))
rl_df = sort_df(pd.read_csv("output_tables/rl_rewards_per_100.csv"))
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)


ax = random_df['averages'].plot(figsize=(40,10), label='Averages')
ax.grid(True)
plt.xlabel('Number of Episodes', fontsize=12)
plt.ylabel('Average rewards per 100 Episodes',  fontsize=12)
ax.legend()

```
